```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import tree, preprocessing
import seaborn as sns
import numpy as np

DF = pd.read_csv("preprocessed_data.csv")
```

```python
DF
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(1234)

train,test = train_test_split(DF,test_size=0.2)
```

```python
from sklearn import preprocessing
def prep_df(data_df):
    
    df = data_df.copy()
    X = df[["ap_hi", "ap_lo","smoke"]]
    y = df[["cardio"]]
    
    return(X, y)
```

```python
X_train, y_train = prep_df(train)
```

```python
X_test, y_test = prep_df(test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
def check_complexity_score(c):

    LR = LogisticRegression(max_iter=100, C = c)
    return cross_val_score(LR,X_train,y_train.values.ravel(),cv=5).mean()
```

```python
x1 = np.logspace(-2, 5, 50, endpoint=True)
```

```python
Complex_score = []
for c in x1:
    x=check_complexity_score(c)
    Complex_score.append((x,c))
```

```python
print(sorted(Complex_score, key=lambda Complex_score: Complex_score[0])[-1])
```

```python
C = sorted(Complex_score, key=lambda Complex_score: Complex_score[0])[-1][1]
```

```python
LR = LogisticRegression(max_iter=1000000, C=C)
# Set LR as the Logistic Regression model with the best C value we've chosen
LR.fit(X_train,y_train.values.ravel())        # Let X_train and y_train fit the model T
LR.score(X_test, y_test)  
```
